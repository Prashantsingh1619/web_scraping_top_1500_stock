```python
final=pd.DataFrame()
for x in range(1,61):
    webpage=requests.get("https://www.screener.in/screens/29729/top-1000-stocks/?page={}".format(x)).text
    soup=BeautifulSoup(webpage,'lxml')
    s=soup.find('div',class_="responsive-holder fill-card-width")
    s_no=[]
    name=[]
    current_price=[]
    price_to_earn=[]
    market_capitalization=[]
    dividend_yield=[]
    Net_Profit_latest_quarter=[]
    YOY_Quarterly_profit_growth=[]
    Sales_latest_quarter=[]
    YOY_Quarterly_sales_growth=[]
    Return_on_capital_employed=[]
    Profit_after_tax_last_year=[]
    
    for i in range(1,27):
        if i==16:
            continue
        s_no.append(s.find_all('tr')[i].find_all('td')[0].text.strip())
        name.append(s.find_all('tr')[i].find_all('td')[1].text.strip())
        current_price.append(s.find_all('tr')[i].find_all('td')[2].text.strip())
        price_to_earn.append(s.find_all('tr')[i].find_all('td')[3].text.strip())
        market_capitalization.append(s.find_all('tr')[i].find_all('td')[4].text.strip())
        dividend_yield.append(s.find_all('tr')[i].find_all('td')[5].text.strip())
        Net_Profit_latest_quarter.append(s.find_all('tr')[i].find_all('td')[6].text.strip())
        YOY_Quarterly_profit_growth.append(s.find_all('tr')[i].find_all('td')[7].text.strip())
        Sales_latest_quarter.append(s.find_all('tr')[i].find_all('td')[8].text.strip())
        YOY_Quarterly_sales_growth.append(s.find_all('tr')[i].find_all('td')[9].text.strip())
        Return_on_capital_employed.append(s.find_all('tr')[i].find_all('td')[10].text.strip())
        Profit_after_tax_last_year.append(s.find_all('tr')[i].find_all('td')[11].text.strip())
    
    page=pd.DataFrame({"s_no":s_no,
    "name":name,
    "current_price(Rs)":current_price,
    "price_to_earn":price_to_earn,
    "market_capitalization(Rs.Cr.)":market_capitalization,
    "dividend_yield(%)":dividend_yield,
    "Net_Profit_latest_quarter(Rs.Cr.)":Net_Profit_latest_quarter,
    "YOY_Quarterly_profit_growth(%)":YOY_Quarterly_profit_growth,
    "Sales_latest_quarter(Rs.Cr.)":Sales_latest_quarter,
    "YOY_Quarterly_sales_growth(%)":YOY_Quarterly_sales_growth,
    "Return_on_capital_employed(%)":Return_on_capital_employed,
    "Profit_after_tax_last_year(Rs.Cr.)":Profit_after_tax_last_year,})
    final = pd.concat([final, page], ignore_index=True)
    print(x)
```
